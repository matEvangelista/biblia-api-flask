```python
import json
```

```python
f = open('biblia-catolica.json')
biblia = json.load(f)
biblia
```

```python
[livro for livro in biblia if livro['livro'] == 'ex' and livro['capitulo'] == 1][0]#['versiculos'][2:10]
```

```python
def get_verse(book: str, chapter: int, verse: int):
    verse = {
        'book': book,
        'chapter': chapter,
        'verse': verse,
        'data': [boo for boo in biblia if boo['livro'] == book and boo['capitulo'] == chapter][0]['versiculos'][
            verse - 1]
    }
    return verse
```

```python
get_verse('mt', 16, 18)
```
